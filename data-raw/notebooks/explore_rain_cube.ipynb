```python
import xarray as xr
# Define where the processed climate data is stored.
CLIMBRA_folder = '../workspace/CLIMBRA/'
# Select which "hazard" (climate risk category) and "indicator" (climate index) to analyze.
hazard = 'ExtremeRainfall'   # Hazard group (rainfall extremes, drought, heat, etc.)
indicator = 'Rx1day'         # Indicator: maximum 1-day precipitation
# Build the full path to the NetCDF file that contains the ensemble of model simulations
file_path = f'{CLIMBRA_folder}/{hazard}/{indicator}/ensemble/ensemble_return_period.nc'
ds = xr.open_dataset(file_path)[indicator]
# Choose how to look at the model ensemble:
ensemble = 'median'
# Select a global warming level (GWL) and return period (years)
GWL = '3'
return_period = 10
# Extract the subset of data for this configuration (median ensemble, 3C GWL, 10-year return period)
ds_ens = ds.sel(ensemble=ensemble, GWL=GWL, return_period=return_period)
# Plot the map of Rx1day values
ds_ens.plot()
```
